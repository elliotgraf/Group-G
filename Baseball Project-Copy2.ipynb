```python
#dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
#create variable to hold csv file
file_path = "baseball_project.csv"
```

```python
# Read our Data file with the pandas library
base_df=pd.read_csv(file_path)
# base_df = base_df.set_index("yearID")
base_df.head()
```

```python
base_df = base_df[base_df["yearID"]>=2010]
hit = base_df.groupby("yearID").mean()["H"]
hit = pd.DataFrame(hit)
hit
```

```python
a = hit.iloc[0:4]
a_mean = a.H.mean()
b = hit.iloc[4:]
b_mean = b.H.mean()
print(a)
print("mean:", a_mean)
print(b)
print("mean:",b_mean)
```

```python
(t_stat, p) = stats.ttest_ind(a, b, equal_var=False)
print("t-statistics is {}.".format(t_stat))
print("p=value is {}.".format(p))
```

```python
x_axis = hit.index
y_axis = hit["H"]
plt.bar(x_axis,y_axis)
```

```python
#create variable to hold csv file
file_path2 = "Copy_of_shift_data_data_table.csv"
```

```python
# Read our Data file with the pandas library
base_df2=pd.read_csv(file_path2)
# base_df = base_df.set_index("yearID")
base_df2
```

```python
renamed_base_df2 = base_df2.rename(columns={"Unnamed: 0":"yearID", "Unnamed: 1":"Team", "Unnamed: 2":"TBF"})

renamed_base_df2.drop(renamed_base_df2.index[0], inplace=True)


renamed_base_df2.yearID = pd.to_numeric(renamed_base_df2.yearID)
renamed_base_df2.TBF = pd.to_numeric(renamed_base_df2.TBF)

type(renamed_base_df2.yearID.iloc[0])
```

```python
renamed_base_df2_new = renamed_base_df2[renamed_base_df2.yearID>2010]
TBF = renamed_base_df2.groupby("yearID").sum()["TBF"]
TBF = pd.DataFrame(TBF)
TBF 
```

```python
# caluclate avg hits(2010 - 2013 and 2014 - 2017)
a = TBF.iloc[0:4]
a_mean = a.TBF.mean()
b = TBF.iloc[4:]
b_mean = b.TBF.mean()
print(a)
print("mean:", a_mean)
print(b)
print("mean:",b_mean)
```

```python
# T-Test of avg RA (2010 - 2013 and 2014 - 2017)
(t_stat, p) = stats.ttest_ind(a, b, equal_var=False)
print("t-statistics is {}.".format(t_stat))
print("p=value is {}.".format(p))
```

```python
x_axis = TBF.index
y_axis = TBF["TBF"]
plt.bar(x_axis,y_axis)

plt.savefig("shift.png")
```
